# INTRODUCTION

General workflow is, Obtaining data and processing it, and then train our ML model on the data, But we also should test it to see if the model is actually trained or just memorised or didnt properly learn from it,
the data is split in two parts, majority (70 to 80 percent) being to train the model and the rest to test the model,
kind of like how your teacher gives you a textbook to prepare for exams but some questions are out of book to really test your understanding of concepts and its application<br> <br>
सामान्य कार्यप्रवाह (General Workflow) में, **डेटा प्राप्त करना और उसका पूर्व-संसाधन (Processing) करना शामिल है**, जिसके बाद हम अपने **ML मॉडल को इस डेटा पर प्रशिक्षित (Train) करते हैं**।  

हालांकि, यह भी आवश्यक है कि हम मॉडल का परीक्षण (Test) करें, ताकि यह सुनिश्चित किया जा सके कि **मॉडल ने वास्तव में सीख लिया है**, न कि केवल डेटा को याद (Memorize) कर लिया है या उसे सही ढंग से समझने में विफल रहा है।  

इसलिए, **डेटा को दो भागों में विभाजित किया जाता है**:  
**70-80% डेटा** मॉडल को प्रशिक्षित (Train) करने के लिए प्रयोग किया जाता है। **शेष डेटा** मॉडल का परीक्षण (Test) करने के लिए उपयोग किया जाता है।  

यह प्रक्रिया **वैसे ही है जैसे एक शिक्षक परीक्षा की तैयारी के लिए एक पाठ्यपुस्तक देता है**, लेकिन **कुछ प्रश्न पाठ्यपुस्तक से बाहर होते हैं**, ताकि यह सुनिश्चित किया जा सके कि छात्र ने अवधारणाओं को वास्तव में समझा है और उनका सही ढंग से अनुप्रयोग कर सकता है।

Dependencies imported<br>
**आवश्यक लाइब्रेरीज़ (Dependencies) आयात की गईं।**  

```python
import pandas as pd
```

Dataset Imported<br>
**डेटासेट आयात किया गया।**  

```python
dataset = pd.read_csv('/content/diabetes.csv')
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
```

We have saperated a column from the dataset, so we have two saperate datasets, one without the column and one itself a column<br>
हमने डेटासेट से एक कॉलम अलग कर दिया है, जिससे अब हमारे पास **दो अलग-अलग डेटासेट** हैं एक **डेटासेट जिसमें वह कॉलम नहीं है**,दूसरा **स्वयं वही कॉलम, एक अलग डेटासेट के रूप में**।

```python
X = dataset.drop('Age', axis = 1)
Y = dataset.Age
```

We have run this block of code to perform train test split, test_size = 0.2 means 20 percent of data is used as test<br>
हमने इस कोड ब्लॉक को **ट्रेन-टेस्ट स्प्लिट (Train-Test Split)** करने के लिए चलाया है।  
**test_size = 0.2** का अर्थ है कि **डेटा का 20% भाग परीक्षण (Test) के लिए उपयोग किया गया है**, जबकि शेष **80% डेटा मॉडल को प्रशिक्षित (Train) करने के लिए** इस्तेमाल किया गया है।  

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```
